```python
import urllib2 
response = urllib2.urlopen('http://www.people.com/people/celebrities/')
html = response.read()
html
```

```python
import re
pattern = r"(?=\<dd\>\<a href\=.*\>)\w+\s\w+(?=\<\/a\>\<\/dd\>)"
pattern = r"(?<=id\=\"\d{8}\"\>)\w+\s\w+"
celebList = re.findall(pattern, html)
print celebList
```
